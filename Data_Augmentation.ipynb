```python
import cv2
from PIL import Image
from PIL import ImageOps
from PIL import ImageFilter
import matplotlib.pyplot as plt


def preprocesspath(image_path):
    patch=cv2.imread(image_path)
    im_as_arr=Image.fromarray(patch)
    return im_as_arr
#image_path=data.path[1]
#preprocesspath(image_path)  
```

```python
def augumentation_image(image):
    augmented_images=[]
   
    augmented_images.append(image)
    
    rotated_image=image.rotate(180)
    augmented_images.append(rotated_image)
    
    mirrored_img=ImageOps.mirror(image)
    augmented_images.append( mirrored_img)
    
    cc=ImageOps.flip(image)
    augmented_images.append( cc)
    
    b = ImageOps.scale(image, factor=0.1, resample=3)
    augmented_images.append(b)
    
    return augmented_images
```

```python
from torchvision import transforms

def plot_augumentationimage(augmented_images):
    
    plt.figure(1,figsize=(20,20))
    n=1
    for image in augmented_images:
        plt.subplot(3,3,n)
        plt.imshow(image)
        
        n=n+1
    plt.show()
```
